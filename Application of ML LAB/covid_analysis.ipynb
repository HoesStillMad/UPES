```python
import pandas as pd        # importing pandas used for data analysis
import numpy as np         # importing numpy helpfull in making n-dimentional arrays
import seaborn as sns      # importing seaborn used for data visualization
import matplotlib.pyplot as plt  # importing matplotlib used for data visualization
```

```python
df=pd.read_csv("covid.csv")   #uploading csv file(dataset) using pandas 
df.head()                     #displaying first five values of dataset
```

```python
df.info()  #checking total values and data type of features
```

```python
df.isnull().sum()   #checking wheather there is any null value in dataset or not
```

```python
df.describe()   #Checking the data
```

```python
sns.pairplot(df) #visualizing the relationship of all fetures with each other
```

```python
df = df.sort_values(by=['Total Cases'], ascending=False)   #Sorting the data according to total cases   
plt.figure(figsize=(15,15))                                #assigining the figure size
plot = sns.barplot(df['Total Cases'], df['State/UTs'])     #plotting a bar garph with two parameters total cases and state names

for i,(value,name) in enumerate(zip(df['Total Cases'],df['State/UTs'])):     #displaying the total number of confirmed cases for each state on the bar graph
    plot.text(value,i,f'{value:,.0f}',size=10)  
plt.show()                           #ploting the visualized bar graph
```

```python
#finding out the correlation between all fetures of dataset

plt.figure(figsize=[20,10])        #assigining the figure size
sns.heatmap(df.corr(),annot=True)  #implementing correlation of dataset on heatmap
```

```python
sns.relplot(x='Total Cases',y='Discharged',data=df)   #plotting a scater plot between total cases and recoverd 
sns.relplot(x='Total Cases',y='Deaths',data=df)       #plotting a scater plot between total cases and Deaths
sns.relplot(x='Total Cases',y='Active',data=df)       #plotting a scater plot between total cases and Active cases
```
